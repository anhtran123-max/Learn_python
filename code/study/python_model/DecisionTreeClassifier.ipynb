```python
import pandas as pd
import warnings
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

import pandas_profiling
from sklearn.metrics import confusion_matrix, accuracy_score
warnings.filterwarnings("ignore")

data_frame = pd.read_csv("housing.csv")
```

```python
data_frame = data_frame.dropna(axis=1)
data_frame
```

```python
print(data_frame.nunique())
```

```python
print(data_frame.shape)
```

```python
y = data_frame.median_house_value	
x = data_frame.drop(["longitude","latitude","median_house_value"],axis=1)
```

```python
x
```

```python
x = data_frame.drop(["ocean_proximity"],axis=1)
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size= 0.2 , random_state= 42)
data_model = DecisionTreeClassifier()
```

```python
pred_train = data_model.fit(x_train, y_train)
true_train = pred_train.predict(x_test)
```

```python
print("Accuracy on train data: ", accuracy_score(true_train , y_test))
```
