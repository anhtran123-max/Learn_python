```python
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split
from sklearn.neighbors import RadiusNeighborsClassifier
import matplotlib.pyplot as plt

import pandas_profiling
from sklearn.metrics import confusion_matrix, accuracy_score
warnings.filterwarnings("ignore")

data_frame = pd.read_csv("housing.csv")
```

```python
data_frame = data_frame.dropna(axis=1)
data_frame
```

```python
y = data_frame.median_house_value	
x = data_frame.drop(["longitude","latitude","median_house_value","ocean_proximity"],axis=1)
```

```python
x
```

```python
# Set Variable To predict
predict =data_frame.median_house_value	
# Set feature which is a component to make platform to predict
data_feature = ['housing_median_age','total_rooms','population','households','median_income']
x = data_frame[data_feature]
```

```python
print(x.describe())
print(x.head())
print(x.value_counts())
print(x)
```

```python
data_model = RadiusNeighborsClassifier()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size= 0.2 , random_state= 42)

pred_train = data_model.fit(x_train ,y_train)
true_train = pred_train.predict(x_train)
# Check probality on Train data
print("Accuracy on train data: ", accuracy_score(true_train , y_train))
```
